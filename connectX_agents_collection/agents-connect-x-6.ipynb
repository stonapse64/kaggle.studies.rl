```python
import copy
import numpy as np
import lightgbm as lgb

class CX_lgbm:
    def __init__(self, agent_name, path_to_model):
        self.name = agent_name
        self.loaded_booster = lgb.Booster(model_file=path_to_model)    
    def fe_board(self, board):
        board2 = copy.deepcopy(board)
        board2 = np.append(board2, [(board == 0).sum()])
        return board2
    def agent(self, observation, configuration):
        board = np.array(observation.board, dtype=int)
        if observation.mark == 2:
            ind1,ind2 = board==1,board==2
            board[ind1],board[ind2] = 2,1
        probs = self.loaded_booster.predict(self.fe_board(board).reshape(1, -1))[0]
        probs *= board[: configuration.columns] == 0
        return int(probs.argmax())
```

```python
from IPython.display import clear_output
from joblib import Parallel, delayed
import os
import time
import random
import pandas as pd
from tqdm import tqdm
from types import SimpleNamespace
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
# import torch
# import shutil
# from plotly import graph_objects as go
```

```python
class CX_gbm:
    def __init__(self, agent_name, gbm):
        self.name = agent_name
        self.gbm = gbm    
    def fe_board(self, board):
        board2 = copy.deepcopy(board)
        board2 = np.append(board2, [(board == 0).sum()])
        return board2
    def agent(self, observation, configuration):
        board = np.array(observation.board, dtype=int)
        if observation.mark == 2:
            ind1,ind2 = board==1,board==2
            board[ind1],board[ind2] = 2,1
        probs = self.gbm.predict(self.fe_board(board).reshape(1, -1))[0]
        probs *= board[: configuration.columns] == 0
        return int(probs.argmax())
    
class CX_ldf:
    def __init__(self, agent_name, df):
        self.name = agent_name
        X_all, y_all = process_df(df)
        lgb_all = lgb.Dataset(X_all, y_all, weight=df["freq"].values)
        self.gbm = lgb.train(params, lgb_all, valid_sets=lgb_all, num_boost_round=500) #, verbose_eval=10) 
    def fe_board(self, board):
        board2 = copy.deepcopy(board)
        board2 = np.append(board2, [(board == 0).sum()])
        return board2
    def agent(self, observation, configuration):
        board = np.array(observation.board, dtype=int)
        if observation.mark == 2:
            ind1,ind2 = board==1,board==2
            board[ind1],board[ind2] = 2,1
        probs = self.gbm.predict(self.fe_board(board).reshape(1, -1))[0]
        probs *= board[: configuration.columns] == 0
        return int(probs.argmax())

def collect_observations_single(agent1, agent2, shuffle=True):
    res = []
    env = create()
    if shuffle and random.choice([True, False]):
        env.run([agent2, agent1])
    else:
        env.run([agent1, agent2])
    for step in env.steps[:-1]:
        board = np.array(step[0].observation.board)
        if (board.sum() == 0) or ((board == 0).sum() == 0):
            continue
        res.append("".join([str(x) for x in board]))
        ind1 = board == 1
        ind2 = board == 2
        board[ind1] = 2
        board[ind2] = 1
        res.append("".join([str(x) for x in board]))
    return res


def collect_observations(agent1, agent2, n_games=1, n_jobs=-1, shuffle=True):
    if n_jobs != 1:
        res = Parallel(n_jobs=-1, temp_folder="/tmp", max_nbytes=None, backend="loky")(
            delayed(collect_observations_single)(agent1, agent2, shuffle)
            for _ in tqdm(range(n_games)))
    else:
        res = [collect_observations_single(agent1, agent2, shuffle) for _ in range(n_games)]
    res = [item for sublist in res for item in sublist]
    return res


def fe(board):
    lst = [int(c) for c in board]
    lst.append((np.array(lst) == 0).sum())
    return lst


def fe_board(board):
    board2 = copy.deepcopy(board)
    board2 = np.append(board2, [(board == 0).sum()])
    return board2


def process_df(df):
    return np.array(df["board"].apply(fe).values.tolist()), df["action"].values


def get_action(ob, configuration, agent):
        observation = SimpleNamespace(**dict(board = [int(x) for x in ob], mark=1))
        return agent(observation, configuration) # , shuffle_actions=False)    

    
def make_df_collect_observations(ags_teacher, ags_opponent, n_games):
    obs = []
    for agt in ags_teacher:
            for ago in ags_opponent:
                obs += collect_observations(agt, ago, n_games)
    print("lena = ", len(obs))
    obs = pd.Series(obs).value_counts().rename("freq").reset_index().rename({"index":"board"}, axis=1)
    env = create()
    actions = Parallel(n_jobs=-1, temp_folder="/tmp", max_nbytes=None)(
        delayed(get_action)(ob, env.configuration, ags_teacher[0]) for ob in tqdm(obs["board"]))
    return pd.DataFrame(dict(board=obs["board"], freq=obs["freq"], action=actions))


params = {
    'boosting_type': 'gbdt',
    'objective': 'multiclass',
    'num_class': create().configuration.columns,
    'num_leaves': 293,
    "min_data_in_leaf": 3,
    'learning_rate': 0.05,
    "feature_pre_filter": False,
    'verbose': -1,
}


def make_agi(ag_name, ags_teacher, ags_opponent, n_Games):
    '''
    ags_teacher = [ag_cell_swarm_rabbie]
    ags_opponent = ["random"] # ["random",'negamax'] # ["random", 'negamax', ag_Sananda_Chowdhury]
    '''
    name_csv   = ag_name + '.csv'
    name_model = ag_name + ".model"

    df_f = make_df_collect_observations(ags_teacher, ags_opponent, n_games=n_Games)

    display(df_f)
    
    X_all, y_all = process_df(df_f)
    lgb_all = lgb.Dataset(X_all, y_all, weight=df_f["freq"].values)
    gbm = lgb.train(params, lgb_all, valid_sets=lgb_all, num_boost_round=1000) #, verbose_eval=10)
    
    gbm.save_model(name_model), df_f.to_csv(name_csv), print (name_model, 'produced.')
```
